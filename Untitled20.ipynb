# Task 3
## Step 1: Load the Dataset

In this step, we import the required libraries and load the Titanic dataset. This dataset contains passenger details such as age, gender, class, and survival status.

```python
import pandas as pd

url = "https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv"
df = pd.read_csv(url)

df.head()
```

### Observation

The dataset has been successfully loaded and contains multiple features that will be used for analysis.

## Step 2: Data Cleaning

In this step, we handle missing values and remove irrelevant columns. The 'Age' column is filled using the mean, and the 'Cabin' column is dropped.

```python
df['Age'].fillna(df['Age'].mean(), inplace=True)
df.drop('Cabin', axis=1, inplace=True)
```

### Observation

Missing values have been handled, and unnecessary columns have been removed. The dataset is now clean and ready for analysis.

# 3 Feature Engineering
## Step 3(a): Create Family Size Feature

We create a new column 'FamilySize' by adding the number of siblings/spouses (SibSp) and parents/children (Parch). This helps analyze the impact of family size on survival.

```python
df['FamilySize'] = df['SibSp'] + df['Parch']
```

```python
df.head()
```

### Observation

The 'FamilySize' column has been successfully created and will be used for further analysis.

## Step 3(b): Create Age Groups

In this step, we divide passengers into different age groups using bins. This helps us understand survival patterns across different age categories.

```python
df['AgeGroup'] = pd.cut(df['Age'], bins=[0,12,20,40,60,100])
```

```python
df.head()
```

### Observation

The 'AgeGroup' column has been created successfully, allowing us to analyze survival rates based on age groups.

# 4 Analysis (Groupby)
## Step 4(a): Survival Rate by Age Group

This step analyzes how survival rates vary across different age groups.

```python
df.groupby('AgeGroup')['Survived'].mean()
```

### Observation

Survival rates differ across age groups. Children tend to have higher survival rates compared to older passengers.

## Step 4(b): Survival Rate by Embarkation Port

This step examines how the embarkation port affected survival chances.

```python
df.groupby('Embarked')['Survived'].mean()
```

### Observation

Passengers from different embarkation ports show variation in survival rates.

## Step 4(c): Survival Rate by Family Size

This step analyzes how family size impacts survival probability.

```python
df.groupby('FamilySize')['Survived'].mean()
```

### Observation

Passengers with smaller family sizes tend to have higher survival rates compared to larger families.

# Visualization
## Step 5(a): Age Distribution

This histogram shows how passenger ages are distributed in the dataset.

```python
import matplotlib.pyplot as plt

plt.hist(df['Age'], bins=20)
plt.show()
```

### Observation

Most passengers fall within the young to middle-age category.

## Step 5(b): Correlation Heatmap

This heatmap shows the correlation between numerical features in the dataset.

```python
import seaborn as sns

sns.heatmap(df.corr(numeric_only=True), annot=True)
plt.show()
```

### Observation

The heatmap helps identify relationships between features such as fare, class, and survival.

## Step 5(c): Survival by Family Size

This bar plot visualizes how survival rate changes with family size.

```python
sns.barplot(x='FamilySize', y='Survived', data=df)
plt.show()
```

### Observation

The graph shows that smaller families had better survival chances.

## Final Conclusion

From this analysis, we conclude that survival on the Titanic was influenced by multiple factors such as age, embarkation port, and family size. Advanced exploratory data analysis helped uncover deeper insights and patterns in the dataset.
